```python
import os
import pandas as pd
import json
import streamlit as st
from streamlit_option_menu import option_menu
import plotly.express as px
```

```python
#Agg_transcation:
state_path="D:/guviproject/phonepe/pulse/data/aggregated/transaction/country/india/state/"
agg_state_trans_list=os.listdir(state_path)


columns_trans={'State': [], 'Years': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],'Transaction_amount': []}

for state in agg_state_trans_list:
    cur_state=state_path + state + "/"
    agg_year_list=os.listdir(cur_state)

    
    for year in agg_year_list:
        cur_year=cur_state + year +"/"
        agg_file_list=os.listdir(cur_year)

        for file in agg_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')

            D1= json.load(data)

            for data in D1['data']['transactionData']:
                    Name=data['name']
                    count=data['paymentInstruments'][0]['count']
                    amount=data['paymentInstruments'][0]['amount']
                    columns_trans['Transaction_type'].append(Name)
                    columns_trans['Transaction_count'].append(count)
                    columns_trans['Transaction_amount'].append(amount)
                    columns_trans['State'].append(state)
                    columns_trans['Years'].append(year)
                    columns_trans['Quarter'].append(int(file.strip('.json')))   

df_agg_trans= pd.DataFrame(columns_trans)

df_agg_trans['State']=df_agg_trans['State'].str.replace("-"," ")
df_agg_trans['State']=df_agg_trans['State'].str.title()
```

```python
# Agg_user:
state_user_path="D:/guviproject/phonepe/pulse/data/aggregated/user/country/india/state/"
agg_state_user_list=os.listdir(state_user_path)

columns_user={'State': [], 'Years': [], 'Quarter': [], 'Brands': [], 'Transaction_count': [],'Percentage': []}


for state in agg_state_user_list:
    cur_state=state_user_path + state + "/"
    agg_year_list=os.listdir(cur_state)

    
    for year in agg_year_list:
        cur_year=cur_state + year +"/"
        agg_file_list=os.listdir(cur_year)


        for file in agg_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')

            D2= json.load(data)
            
            try:
                for data in D2["data"]["usersByDevice"] :
                        brands= data["brand"]
                        counts = data["count"]
                        percent = data["percentage"]
                        columns_user["Brands"].append(brands)
                        columns_user["Transaction_count"].append(counts)
                        columns_user["Percentage"].append(percent)
                        columns_user["State"].append(state)
                        columns_user["Years"].append(year)
                        columns_user["Quarter"].append(int(file.strip(".json")))
            except:
                 pass

df_agg_user=pd.DataFrame(columns_user)

df_agg_user['State']=df_agg_user['State'].str.replace("-"," ")
df_agg_user['State']=df_agg_user['State'].str.title()
```

```python
# MAP TRANSACTION:
map_trans_path="D:/guviproject/phonepe/pulse/data/map/transaction/hover/country/india/state/"
map_trans_list=os.listdir(map_trans_path)

columns_map_trans = {'State': [], 'Years': [], 'Quarter': [], 'District': [], 'Count': [],'Amount': []}

for state in map_trans_list:
    cur_state=map_trans_path + state + "/"
    map_year_list=os.listdir(cur_state)

    
    for year in map_year_list:
        cur_year=cur_state + year +"/"
        map_file_list=os.listdir(cur_year)


        for file in map_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')

            D4= json.load(data)

            for data in D4["data"]["hoverDataList"]:
                district = data["name"]
                count = data["metric"][0]["count"]
                amount = data["metric"][0]["amount"]
                columns_map_trans["District"].append(district)
                columns_map_trans["Count"].append(count)
                columns_map_trans["Amount"].append(amount)
                columns_map_trans['State'].append(state)
                columns_map_trans['Years'].append(year)
                columns_map_trans['Quarter'].append(int(file.strip('.json')))

df_map_trans=pd.DataFrame(columns_map_trans)

df_map_trans['State']=df_map_trans['State'].str.replace("-"," ")
df_map_trans['State']=df_map_trans['State'].str.title()
```
